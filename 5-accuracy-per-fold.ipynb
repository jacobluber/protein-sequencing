```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)

SMALL_SIZE = 12
MEDIUM_SIZE = 12
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

fig, axs = plt.subplots(3, len(all_dfs), figsize=(19,8))
def plot_length(length_range, df_dict=all_dfs):
  for ax in axs.ravel():
    ax.clear()
  
  i=0
  for key,df in df_dict.items():
    subdf = df[(df['seq_length'] >= length_range[0]) & (df['seq_length'] <= length_range[1])]
    lengths = subdf['seq_length']
    matching_scores = subdf['Matching Percentage']
    
    axs[0,i].scatter(lengths, matching_scores, marker='o', s=2, alpha = 0.1)
    axs[0,i].set_xlabel("Length of sequences [residues]")
    axs[0,0].set_ylabel("Unmasking accuracy [%]")
    axs[0,i].set_xlim(length_range)
    axs[0,i].set_title(key,fontsize=BIGGER_SIZE)
    
    axs[1,i].hist(lengths, bins = range(length_range[0],length_range[1],50), alpha=0.7, edgecolor='grey')
    axs[1,i].set_xlabel("Length of sequences [residues]")
    axs[1,0].set_ylabel("Density [counts]")
    axs[1,i].set_xlim(length_range)
    
    n , _ , _ = axs[2,i].hist(matching_scores, bins = range(-1,100,5), alpha=0.7, edgecolor='grey', orientation='horizontal')
    axs[2,i].set_xlabel("Density [counts]")
    axs[2,0].set_ylabel("Unmasking accuracy [%]")
    axs[2,i].set_ylim([0,100])
    average_score = np.mean(matching_scores)
    axs[2,i].text(max(n), 0, f'Avg: {average_score:.2f}%', fontsize=SMALL_SIZE, color='black', ha='right', va='bottom')


    i+=1
    
  plt.tight_layout()

  plt.close(fig)
  return fig

plot_length([0,1500])
```
